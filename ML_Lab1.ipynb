```python
dataset = []

with open("adultIncomeCensus.csv", "r") as file:
    for line in file:
        row = line.strip().split(",")
        dataset.append(row)

print("Total records loaded:", len(dataset))
```

```python
print("10th record:")
print(dataset[9])

print("19th record from last:")
print(dataset[-19])
```

```python
attributes = len(dataset[0])
print("Number of attributes per record:", attributes)
```

```python
age = 0
education = 3
occupation = 6
hours_per_week = 12
```

```python
ages = []
education_levels = []
occupations = []
hours_list = []

for row in dataset:
    ages.append(row[age])
    education_levels.append(row[education])
    occupations.append(row[occupation])
    hours_list.append(row[hours_per_week])

print("Lists created successfully!")
```

```python
cleaned_data = []

for row in dataset:
    if "?" not in row:
        cleaned_data.append(row)

print("Original records:", len(dataset))
print("Cleaned records:", len(cleaned_data))
```

```python
first_500 = cleaned_data[:500]
last_200 = cleaned_data[-200:]

combined_subset = first_500 + last_200

print("Combined subset size:", len(combined_subset))
```

```python

for row in cleaned_data:
    if row[14] == "<=50K":
        row[14] = "Low"
    elif row[14] == ">50K":
        row[14] = "High"

print("Income labels updated!")
```

```python
replicated_data = cleaned_data[:100] * 5

print("Replicated dataset size:", len(replicated_data))
```
